```python
import requests
from bs4 import BeautifulSoup
import csv
import time

def google_search(query, num_pages):
    urls = []
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko)Chrome/91.0.4472.124 Safari/537.36"
    }
    base_url = "https://www.google.com/search?q=" + query + "&start="

    for page in range(num_pages):
        start = page * 10
        url = base_url + str(start)
        response = requests.get(url, headers=headers)
        if response.status_code != 200:
            print(f"Failed to retrieve page {page + 1}")
            continue

        soup = BeautifulSoup(response.text, 'html.parser')
        search_results = soup.find_all('div', class_='g')

        for result in search_results:
            link_tag = result.find('a', href=True)
            if link_tag:
                href = link_tag['href']
                if "arcgis/rest/services" in href:
                    urls.append(href)

        # Be respectful to Google's server
        time.sleep(2)

    return urls

def save_to_csv(urls, filename="arcgis_urls.csv"):
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(["URL"])
        for url in urls:
            writer.writerow([url])

if __name__ == "__main__":
    query = "arcgis/rest/services"
    num_pages = 78  # Number of pages to scrape
    urls = google_search(query, num_pages)
    save_to_csv(urls)
    print(f"Extracted {len(urls)} URLs and saved to 'arcgis_urls.csv'")
```
